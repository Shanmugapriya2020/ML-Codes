## If you want to implement Linear Regression, target variable should be continuous, features could be cont or categorical

## Not allowed target variable is categorical

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

```python
df=pd.read_csv("./headbrain.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
X=df['Head Size(cm^3)'].values
```

```python
Y=df['Brain Weight(grams)'].values
```

macine larning, X should be 2 dimensionl

```python
X=X.reshape(-1,1)
```

```python
X.ndim
```

```python
#order matters
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,Y,train_size=0.20)
```

```python
X_test
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)# run the model
```

```python
y_pred=lr.predict(X_test)
```

```python
y_pred
```

```python
r2=r2_score(y_test,y_pred)
```

```python
r2
```

```python
plt.scatter(X_test, y_test,color='black',label='Actual')
plt.plot(X_test, y_pred,color='red',label='Predicted')

plt.legend()
```
